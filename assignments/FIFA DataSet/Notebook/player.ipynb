## Preprocessing

```python
# Common attribute
def calculate_age(age_val):
    year, days_into_season = age_val.split(chr(45))
    year, days_into_season = int(year), int(days_into_season)
    start_date = "2022-11-20" # Qatar worldcup 2022 start date
    start_date = datetime.strptime(start_date, "%Y-%m-%d")

    # Calculate the age value
    age_value = year + ((start_date - datetime(start_date.year, 1 if start_date.month >= 8 else 2, 1)).days + days_into_season) / 365.25
    return age_value

# Derive at which field each player tackled
# To be displayed at "divergent"
def position_at_which_player_tackled_most(defense, mid, att):

    if np.isnan(defense) or np.isnan(mid) or np.isnan(att):
        res = None
    else:
        defense, mid, att = int(defense), int(mid), int(att)
        total_tackles = defense + mid + att
        if total_tackles == 0:
            res = None
        else:
            ratio_def = defense/total_tackles
            ratio_mid = mid/total_tackles
            ratio_att = att/total_tackles

            res = (ratio_def*1 + ratio_mid*2 + ratio_att*3)/3
            # If res closer to 1, consider att tackler
            # elif closre to 0.4 and 0.5, balanced contribution accross three fieldss
            # elif res closer to 0.3 consider def field tackler
    
    return res
```

```python
df_player_defense['calculated_age'] = [calculate_age(age) for age in df_player_defense['age']]
df_player_gca['calculated_age'] = [calculate_age(age) for age in df_player_gca['age']]
df_player_keepers['calculated_age'] = [calculate_age(age) for age in df_player_keepers['age']]
df_player_keepersadv['calculated_age'] = [calculate_age(age) for age in df_player_keepersadv['age']]
df_player_misc['calculated_age'] = [calculate_age(age) for age in df_player_misc['age']]
df_player_passing['calculated_age'] = [calculate_age(age) for age in df_player_passing['age']]
df_player_passing_types['calculated_age'] = [calculate_age(age) for age in df_player_passing_types['age']]
df_player_playingtime['calculated_age'] = [calculate_age(age) for age in df_player_playingtime['age']]
df_player_possession['calculated_age'] = [calculate_age(age) for age in df_player_possession['age']]
df_player_shooting['calculated_age'] = [calculate_age(age) for age in df_player_shooting['age']]
df_player_stats['calculated_age'] = [calculate_age(age) for age in df_player_stats['age']]
```
